`A perceptron is the simplest type of artificial neural network used for binary classification, inspired by the function of a biological neuron.`

**Definition:**

A perceptron is a computational model that mimics the decision-making process of a neuron in the human brain. It is primarily used for binary classification tasks, where it predicts one of two possible outcomes based on input features. Introduced by Frank Rosenblatt in 1957, the perceptron was designed to learn from examples and adjust its parameters to improve accuracy in classifying new data.

**Sigmoid Function Perceptron :**

`The sigmoid function perceptron is a type of artificial neuron that uses the sigmoid function as its activation function`




**random.randn():** `Generates random numbers from a standard normal distribution`
That means:

Mean = 0

Standard deviation = 1

Values can be positive or negative (e.g., -0.5, 1.2, etc.)
**input_size:** `Defines how many random numbers you want`

If input_size = 3, it returns something like :
array([ 0.23, -1.45, 0.67 ])

```python
class Sigmoid():
  def __init__(self,input_size):
    self.weight=np.random.randn(input_size)
    self.bias=np.random.randn(1)

  def sigmoid(self,z):
    return 1/(1+np.exp(-z))

  def predict(self,inputs):
    z=np.dot(inputs,self.weight)+self.bias
    return self.sigmoid(z)

  def fit(self,inputs,targets,learning_rate,num_epochs):
    number_example=len(inputs)

    for epoch in range(num_epochs):
      for i in range(number_example):
        input=inputs[i]
        target=targets[i]
        prediction=self.predict(input)
        error=target-prediction

        grediant_weight=error*prediction*(1-prediction)*input
        self.weight+=grediant_weight*learning_rate

        grediant_bias=error*prediction*(1-prediction)
        self.bias+=grediant_bias*learning_rate

  def accuracy(self,inputs,targets):
    correct=0
    for input,target in zip(inputs,targets):
      pred=self.predict(input)
      if(pred>0.5):
        pred=1
      else:
        pred=0
      if(pred==target):
        correct+=1
    return correct/len(inputs)
```

```python

# AND gate dataset
X = np.array([
    [0, 0],
    [0, 1],
    [1, 0],
    [1, 1]
])

y = np.array([0, 0, 0, 1])

model = Sigmoid(input_size=2)

model.fit(X,y,0.1,1000)

for x in X:
    print(f"Input: {x}, Prediction: {model.predict(x)}")
```

```python
acc = model.accuracy(X, y)
print("Accuracy:", acc)
```
