`A perceptron is the simplest type of artificial neural network used for binary classification, inspired by the function of a biological neuron.`

**Definition:**
## A perceptron is a computational model that mimics the decision-making process of a neuron in the human brain. It is primarily used for binary classification tasks, where it predicts one of two possible outcomes based on input features. Introduced by Frank Rosenblatt in 1957, the perceptron was designed to learn from examples and adjust its parameters to improve accuracy in classifying new data.

```python
import numpy as np
```

**Sigmoid Function Perceptron :**
### The **sigmoid function perceptron** is a type of artificial neuron that uses the sigmoid function as its activation function

```python

class  SigmoidPerceptron():
  def __init__(self,input_size):
    self.weights=np.random.randn(input_size)
    self.bias=np.random.randn(1)

  def sigmoid(self,z):
    return 1/(1+np.exp(-z))

  def predict(self,inputs):
    weighted_sum=np.dot(self.weights,inputs)+self.bias
    return self.sigmoid(weighted_sum)

  def fit(self,inputs,targets,learning_rate=0.01,epochs=100):
    num_example=len(inputs)

    for epoch in range(epochs):
      for i in range(num_example):

        input_data=inputs[i]
        target=targets[i]
        prediction=self.predict(input_data)
        error=target-prediction

        ##Udpading the gradient Weight
        grediant_weight=error*prediction*(1-prediction)*input_data
        self.weights+=learning_rate*grediant_weight

        ##Udpading the bias
        grediant_bias=error*prediction*(1-prediction)
        self.bias+=learning_rate*grediant_bias

  def evaluate(self,inputs,targets):
    correct=0

    for input_vector,target in zip(inputs,targets):
      prediction=self.predict(input_vector)
      if(prediction>=0.5):
        prediction=1
      else:
        prediction=0

      if(prediction==target):
        correct+=1
    accuracy=correct/len(inputs)
    return accuracy
```
